# Pr6-2: 色空間の変換


## グレースケール画像化

cv2.cvtColor 関数で RGB から Gray に変えることができます．

```
I_gray = cv2.cvtColor(I_rgb, cv2.COLOR_RGB2GRAY)
```

以下の画像をグレースケール化してみましょう．
行列のサイズはどう変わるでしょうか？


````
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 画像ファイル名の指定
image_file = "slides.png"

# 画像ファイルの読み込み
I_bgr = cv2.imread(image_file)

# BGRからRGBへの変換
I_rgb = cv2.cvtColor(I_bgr, cv2.COLOR_BGR2RGB)

# Work: RGBからGrayに変換してみましょう．
I_gray = cv2.cvtColor(I_rgb, cv2.COLOR_RGB2GRAY)

print("I_rgb.shape: ", I_rgb.shape)
print("I_gray.shape: ", I_gray.shape)
````

### グレースケール画像の表示

グレースケール画像は plt.imshow を調整する必要があります．<br>
以下の違いを確認しておきましょう．

```
plt.imshow(I_gray)

plt.imshow(I_gray, cmap="gray")
```

cmap は何のためにあるのでしょう？


````
# 表示・出力処理
fig = plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(I_rgb)
plt.axis('off')
plt.title("RGB")

plt.subplot(1, 3, 2)
plt.imshow(I_gray)
plt.axis('off')
plt.title("Gray")

plt.subplot(1, 3, 3)

# Work: cmap = "gray"を追加で入れてみましょう．
plt.imshow(I_gray, cmap="gray")
plt.axis('off')
plt.title("Gray (cmap)")
````

## CIE L\*a\*b\*色空間の可視化

CIE L\*a\*b\*色空間（以下 Lab と略記）を可視化してみます．

-   conv_channels に可視化用の Lab のチャンネル名を設定します．
-   RGB から Lab に変換します．

```
I_conv = cv2.cvtColor(I_rgb, cv2.COLOR_RGB2Lab)
```

I_rgb[:,:,i]のようにすると，個別のチャンネルの取り出しができます．

```
# Rチャンネルのプロット
plt.imshow(I_rgb[:,:,0], cmap="gray")

# Gチャンネルのプロット
plt.imshow(I_rgb[:,:,1], cmap="gray")

# Bチャンネルのプロット
plt.imshow(I_rgb[:,:,2], cmap="gray")
```

対応して，Lab も L, a, b の順にデータが入っているので以下で個別のチャンネルを表示できます．

```
# Lチャンネルのプロット
plt.imshow(I_lab[:,:,0], cmap="gray")

# aチャンネルのプロット
plt.imshow(I_lab[:,:,1], cmap="gray")

# bチャンネルのプロット
plt.imshow(I_lab[:,:,2], cmap="gray")
```


````
rgb_channels = ["R", "G", "B"]

# Lab色空間に合わせてチャンネル名を編集します．
conv_channels = ["L", "a", "b"]

# RGBからLabに変換
I_lab = cv2.cvtColor(I_rgb, cv2.COLOR_RGB2Lab)

# 表示・出力処理
fig = plt.figure(figsize=(12, 9))
plt.subplots_adjust(left=0.05, right=0.95, bottom=0.05, top=0.95, hspace=0.2, wspace=0.05)

plt.subplot(3, 3, 1)
plt.imshow(I_rgb)
plt.axis('off')

for i, ch in enumerate(rgb_channels):
    plt.subplot(3, 3, i+4)

    # RGBチャンネルを個別に取り出して表示
    plt.imshow(I_rgb[:, :, i], cmap="gray")

    plt.title("{0}".format(ch))
    plt.axis('off')

for i, ch in enumerate(conv_channels):
    plt.subplot(3, 3, i+7)

    # Labチャンネルを個別に取り出して表示
    plt.imshow(I_lab[:, :, i], cmap="gray")

    plt.title("{0}".format(ch))
    plt.axis('off')
````
