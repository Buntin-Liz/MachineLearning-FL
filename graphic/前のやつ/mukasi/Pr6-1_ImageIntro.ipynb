## まずは読み込みをやってみましょう

1. 左のフォルダUIに画像ファイル読み込んでください（ドラッグ＆ドロップでOK）．

```

```

画像は正しく表示されているでしょうか？

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 画像ファイル名の指定
image_file = "slides.png"

# Work: 画像ファイルを読み込むように書き換えましょう．
I = np.zeros((16, 16, 3))
I = cv2.imread(image_file)
# 画像サイズの出力
print("- image_size: {0}".format(I.shape))

# 表示・出力処理
fig = plt.figure(figsize=(6,6))

plt.imshow(I)
plt.axis('off');
````

## 色を正しく表示してみましょう

OpenCVでは，読み込み時にデフォルトの色空間がBGRの順番になっていて，
MatplotlibでRGBで表示しようとすると色が逆転してしまいます．

- cv2.cvtColor関数でBGRをRGBに変えることができます．


```
I_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
```




````python
# Work: BGRをRGBに変換するように書き換えましょう．
I_rgb = I

# 表示・出力処理
fig = plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(I)
plt.axis('off')
plt.title("BGR")
I_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 2)
plt.imshow(I_rgb)
plt.axis('off')
plt.title("RGB")
````
